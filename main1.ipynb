```python
import pandas as pd
import ipywidgets as wid
import matplotlib.pyplot as plt

communalService = pd.read_csv('communalService.csv', sep=";")
employee = pd.read_csv('employee.csv', sep=';')
employeeOrder = pd.read_csv('employeeOrder.csv', sep=';')
technicsOrder = pd.read_csv('technicsOrder.csv', sep=";")

costGas = 55
costCategory = [100, 125, 140, 150]


def get_category(id_workers):
    category = employee[employee["ID"] == id_workers]["numberCategory"].values[0]
    # employee.loc[employee["ID"] == id_workers]["CostHour"] = costCategory[category-1]
    return costCategory[category-1]

for xx in employeeOrder["employeeNumber"]:
    category =  get_category(xx)
    employeeOrder.loc[employeeOrder["employeeNumber"] == xx, "totalCostEmployee"] = (category * employeeOrder["hours"])
totalCost = employeeOrder["totalCostEmployee"].sum()

display(employeeOrder)
print(f'Total cost {totalCost} Грн')
groupTable = pd.concat([employeeOrder, technicsOrder], ignore_index=True).fillna(0)
# newTable = pd.concat([employeeOrder, technicsOrder])
# display(newTable)
groupTable["totalCostPetrol"] = groupTable['petrol']*costGas
groupTable["totalCostTechnicsAndEmployee"] = groupTable["totalCostPetrol"] + groupTable["totalCostEmployee"]
display(groupTable)


# prices_df = pd.DataFrame(columns=['Computer ID', 'OS ID','Price', 'Amount', 'Total price'])
```
